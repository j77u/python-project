```python
import pandas as pd
df = pd.read_csv('Car_Prices_Poland.csv')
df
```

```python
import matplotlib.pyplot as plt #this is imported so that we can plot a graph
# to see the relations between the price and other factors that could affect it.
%matplotlib inline
```

```python
plt.scatter(df['mileage'],df['price']) #first we check the relation between the mileage and the price
```

```python
plt.scatter(df['age'],df['price'])
# next we check the age and the price.
# we use the price in both cases as it is the output set.
# both the Age and the Mileage have a similar impact on the price, therefore they are the input sets
```

```python
x=df[['mileage','age']]
x # I create an array here which includes both age and mileage as the input set.
```

```python
y=df['price']
y # here is the output set, in this case only the Price
```

```python
from sklearn.model_selection import train_test_split # this is needed to train the AI using the arrays I made earlier.
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
# we will use 80% of the data from the table as the training set.
# the remaining 20% will be used for the test set.
```

```python
#x is part of the input set, which are Mileage and Age.
#here we have the testing set, which will be used to predict the output for the price of these rows.
x_test
```

```python
#y is the output set, the Car Price.
#the results from x_test will be compared to these values to determine the AI's accuracy.
y_test
```

```python
#this is the training set of the input set.
x_train
```

```python
#this is the training set of the output set.
y_train
```

```python
#Linear regression uses the plotted graph to draw a straight line through the dots to predict future values.
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(x_train, y_train)
```

```python
clf.predict(x_test) #this set of results was predicted by the AI based on the data I trained it with.
```

```python
#compare it to the y_test
y_test
```

```python
clf.score(x_test,y_test)
#this is to show the accuracy of the results.
#as we can see, the AI results have a 75% accuracy to the actual results from the table.
```
